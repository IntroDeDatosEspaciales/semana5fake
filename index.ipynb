```python
# read web table into pandas DF
import pandas as pd

forestDFs=pd.read_html(ciaLink2, # link
                        header=0, # where is the header?
                        flavor='bs4')
```

```python
# check object type 
type(forestDFs)
```

```python
# check size
len(forestDFs)
```

```python
# make a copy
forest=forestDFs[0].copy()
```

```python
forest
```

```python
forest.columns
```

```python
forest.columns.to_list()
```

```python
forest['Date of Information'].value_counts()
```

```python
#1 Replace '%' by 'pct'.
forest.columns=forest.columns.str.replace(pat=r'%',repl='pct',regex=True)
forest
```

```python
#2 mantener el nombre de columnas y quitar el rank
forest=forest.drop(columns=['Rank'])
forest
```

```python
#3
forest=forest.rename(columns={"Date of Information": "forest_date"})
forest
```

```python
#4 no debe existir espacio entre el nombre de las columnas
forest.columns=forest.columns.str.replace(pat=r' ',repl='',regex=True)
forest
```

```python
#5
forest['Country']=forest['Country'].str.strip()
forest
```

```python
#6
forest.forest_date.str.extract(pat=r'(\d+)',expand=True)
```

```python
carbon
```

```python
import pandas as pd
carbon=pd.DataFrame(carbon)
```

```python
carbon=carbon.drop(columns=["slug", "ranking"])
carbon
```

```python
# Change the column name date_of_information to carbon_date
carbon=carbon.rename(columns={"date_of_information": "carbon_date"})
carbon
```

```python
#Make sure the cells with text does not have neither trailing nor leading spaces
carbon['name'] = carbon['name'].str.strip()
carbon
```

```python
# Detect the presence of symbols in the numeric data that are not numeric or point
carbon.value.str.contains(pat=r'\D',regex=True)
```

```python
#Keep only the year value in the column carbon_date
carbon.carbon_date.str.extract(pat=r'(\d+)',expand=True)
```
